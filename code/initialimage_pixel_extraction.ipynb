```python
import numpy as np 
import matplotlib.pyplot as plt 

data = np.load('images.npy',mmap_mode='r') #read the file
```

```python
print(data.shape) # number of images
```

```python
plt.imshow(data[6]) #load the image
plt.axis('off') 
plt.show()
```

```python
from PIL import Image 

image = data[0] 
pil_image = Image.fromarray(image.astype('uint8')) #convert into pil format
pil_image.show() 
pixel_values = list(pil_image.getdata())
pil_image.save('my_image.png')
```

```python
np_frmat = np.array(pixel_values) 
width,height = pil_image.size  
reshaped_np_frmat = np_frmat.reshape(width,height,3) 
```

```python
import pandas as pd 

df = pd.DataFrame(pixel_values, columns=['Red', 'Green', 'Blue']) 
df.to_csv('csvformat.csv', index=False)
```

```python
new_df=pd.read_csv('csvformat.csv')
```

```python
# Scatter plot of RGB values ==> 2D plot
plt.figure(figsize=(12, 12))
plt.scatter(new_df['Red'], new_df['Green'], c='green', alpha=0.5, label='Green Channel')
plt.scatter(new_df['Red'], new_df['Blue'], c='blue', alpha=0.5, label='Blue Channel')
plt.scatter(df['Green'], new_df['Blue'], c='red', alpha=0.5, label='Red Channel')
plt.title('RGB Scatter Plot')
plt.xlabel('Red Intensity')
plt.ylabel('Green/Blue Intensity')
plt.legend()
plt.grid()
plt.show()
```

```python

import matplotlib.pyplot as plt
import pandas as pd

# Assuming new_df is your DataFrame with RGB values

plt.figure(figsize=(15, 5))

# Histogram for the Red channel
plt.subplot(1, 3, 1)
plt.hist(new_df['Red'], bins=256, color='red', alpha=0.7)
plt.title('Red Channel Distribution')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency of Intensity')

# Histogram for the Green channel
plt.subplot(1, 3, 2)
plt.hist(new_df['Green'], bins=256, color='green', alpha=0.7)
plt.title('Green Channel Distribution')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency of Intensity')

# Histogram for the Blue channel
plt.subplot(1, 3, 3)
plt.hist(new_df['Blue'], bins=256, color='blue', alpha=0.7)
plt.title('Blue Channel Distribution')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency of Intensity')

# Adjust layout to prevent overlap
plt.tight_layout()

plt.savefig('RGB_channel_histograms.png')
print("File saved successfully!")

# Show the plot
plt.show()

```
